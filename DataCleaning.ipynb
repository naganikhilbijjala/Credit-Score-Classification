```python
import boto3
import pandas as pd
from io import BytesIO
```

```python
bucket_name = 'credit-score-classifiaction-data'
s3 = boto3.client('s3')
response = s3.get_object(Bucket=bucket_name, Key='train.csv')
csv_content = response['Body'].read()
```

```python
df = pd.read_csv(BytesIO(csv_content), low_memory=False)
```
